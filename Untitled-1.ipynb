```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Load data
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

# Assuming the target column is the last column in train_df
X_train = train_df.iloc[:, :-1]
y_train = train_df.iloc[:, -1]
```

```python
# Identify categorical and numerical columns
categorical_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f"Categorical columns: {categorical_cols}")
print(f"Numerical columns: {numerical_cols}")
```

```python
# Create preprocessing steps for categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'  # Keep numerical columns as they are
)

# Create a pipeline with preprocessing and model
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
# Encode target if it's categorical
if y_train.dtype == 'object':
    label_encoder = LabelEncoder()
    y_train = label_encoder.fit_transform(y_train)
    print(f"Target classes: {label_encoder.classes_}")
```

```python
# Train the model
pipeline.fit(X_train, y_train)
```

```python
# Cross-validation
cv_scores = cross_val_score(pipeline, X_train, y_train, cv=5)
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean CV score: {cv_scores.mean():.4f}")
```

```python
# Get feature names after one-hot encoding
if categorical_cols:
    ohe = pipeline.named_steps['preprocessor'].named_transformers_['cat']
    cat_feature_names = ohe.get_feature_names_out(categorical_cols)
    feature_names = np.append(cat_feature_names, numerical_cols)
else:
    feature_names = X_train.columns
```

```python
# Feature importance
feature_importance = pd.DataFrame({
    'Feature': feature_names,
    'Importance': pipeline.named_steps['classifier'].feature_importances_
}).sort_values('Importance', ascending=False)

print("\nFeature Importance:")
print(feature_importance.head(10))  # Top 10 features
```

```python
# Make predictions on test data
if test_df.shape[1] == train_df.shape[1]:
    # If test has labels
    X_test = test_df.iloc[:, :-1]
    y_test = test_df.iloc[:, -1]
    
    # Encode test target if needed
    if y_test.dtype == 'object' and y_train.dtype != 'object':
        y_test = label_encoder.transform(y_test)
    
    y_pred = pipeline.predict(X_test)
    
    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nModel accuracy on test data: {accuracy:.4f}")
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.savefig('confusion_matrix.png')
    
    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
else:
    # If test doesn't have labels
    X_test = test_df
    y_pred = pipeline.predict(X_test)
    
    # Convert predictions back to original labels if needed
    if hasattr(locals(), 'label_encoder'):
        y_pred = label_encoder.inverse_transform(y_pred)
    
    # Save predictions
    predictions_df = pd.DataFrame({'predictions': y_pred})
    predictions_df.to_csv('predictions.csv', index=False)
    print("Predictions saved to 'predictions.csv'")
```

```python
# Plot feature importance (top 10 features)
plt.figure(figsize=(10, 6))
top_features = feature_importance.head(10)
sns.barplot(x='Importance', y='Feature', data=top_features)
plt.title('Top 10 Feature Importance')
plt.tight_layout()
plt.savefig('feature_importance.png')

print("\nResults visualizations saved as 'confusion_matrix.png' and 'feature_importance.png'")
# Calculate training accuracy
train_pred = pipeline.predict(X_train)
train_accuracy = accuracy_score(y_train, train_pred)
print(f"\nTraining accuracy: {train_accuracy:.4f}")

# Make predictions on test data if it has labels
if test_df.shape[1] == train_df.shape[1]:
    X_test = test_df.iloc[:, :-1]
    y_test = test_df.iloc[:, -1]
    
    # Encode test target if needed
    if y_test.dtype == 'object' and y_train.dtype != 'object':
        y_test = label_encoder.transform(y_test)
    
    # Calculate test accuracy
    test_pred = pipeline.predict(X_test)
    test_accuracy = accuracy_score(y_test, test_pred)
    print(f"Test accuracy: {test_accuracy:.4f}")
else:
    print("Test set doesn't have labels, can't calculate accuracy.")
```
